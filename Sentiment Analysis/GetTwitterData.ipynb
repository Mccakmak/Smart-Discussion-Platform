```python
import csv
import snscrape.modules.twitter as sntwitter
import pandas as pd
```

```python
def getTweets(key, since, until, filename):
  
    maxTweets = 100000
    filename = "data/" + filename + '.csv'
    csvFile = open(filename, "w", newline='', encoding='utf8')
    csvWriter = csv.writer(csvFile)
    csvWriter.writerow(['comment_text'])
    #since = "2021-01-1"
    #until = "2021-01-7"

    # Write tweets into the csv file
    for i,tweet in enumerate(sntwitter.TwitterSearchScraper(key +" lang:en since:" + since + " until:" + until).get_items()):
            if i > maxTweets:
                break
            csvWriter.writerow([tweet.content])
    csvFile.close()

    file = pd.read_csv(filename, encoding='latin-1')
    before_drop = file.shape[0]
    file = file.drop_duplicates(subset=['comment_text'])
    after_drop_dub = file.shape[0]

    print("Dublicate dropped rows: " + str(before_drop - after_drop_dub ))
```
